## Maps

### Using Geopandas and Matplotlib

We will now link our data set with a SHP file. We will use Geopandas to manipulate a geodataframe.

```python
import geopandas as gpd

gpd.__version__
```

Let's create our geodataframe by reading in a shapefile.

```python
thai_gdf1 = gpd.GeoDataFrame.from_file('GIS/THA_adm1.shp')

thai_gdf1
```

As you can see above, a GeoPandas dataframe, or geodataframe, is very much like a Pandas dataframe. One main difference is a GeoPandas dataframe can store geospatial data - look at the `geometry` column. 

Let's do a merge of our `mtct3_series_df` dataframe with the `gdf` geodataframe. Note a different syntax for merging a Pandas DataFrame with a Geopandas GeoDataFrame.

### Question

#### How does the `merge` function differ between Pandas and GeoPandas?

```python
mtct_gdf1 = thai_gdf1.merge(mtct_series_df3, on='ISO_CODE', how='inner')
```

```python
mtct_gdf1['TRpct'].max()
```

Now, let's draw our first map using MatPlotLib.

```python
import matplotlib as mpl

years = [2013,2014,2015,2016]

kw = dict(
    vmin=mtct_gdf1['TRpct'].min(),
    vmax=mtct_gdf1['TRpct'].max(),
    legend=False,
    alpha=1.0,
    cmap='coolwarm',
    linewidth=0.2,
    zorder=5,
)

figsize = (18, 5)
```

#### Figure 7

```python
# For the 4 subplots

f, ax = plt.subplots(nrows=1, ncols=4, figsize=figsize, sharey=True, sharex=True)

for i, year in enumerate(years):
    df = mtct_gdf1[mtct_gdf1['year']==year]
    df.to_crs(epsg=4326)
    ax[i].set_title=year
    df.plot(column='TRpct', ax=ax[i], **kw)
    x_axis = ax[i].axes.get_xaxis()
    x_axis.set_label_text(year, fontsize=14)
    
f.suptitle('Thailand Maternal to Child Transmission Rates, 2013-2016', fontsize=16)

# For the color bar 
sm = plt.cm.ScalarMappable(cmap='coolwarm', \
                        norm=plt.Normalize(vmin=0, \
                        vmax=mtct_gdf1['TRpct'].max()))
sm._A = []
cbar_ax, kww = mpl.colorbar.make_axes([axx for axx in ax.flat])
f.colorbar(sm, cax=cbar_ax)
cb = plt.colorbar(sm, cax=cbar_ax, **kww)
cb.set_label("Percent Transmission Rate")
```

### Four Maps, Four Years

Let's make sure the colorbar is normalized across 4 maps. This is a quick and dirty plot of 4 maps for 4 years in series.

```python
import pysal as ps

ps.__version__
```

```python
import pysal as ps
from branca.utilities import split_six

def get_threshold_scale(classifier, num_classes, gdf_series):
    classifier = 'Fisher_Jenks'
    if num_classes > 6:
        num_classes = 6
    gdf_series = gdf_series
    if classifier == 'Fisher_Jenks':
        threshold_scale=ps.esda.mapclassify.Fisher_Jenks(gdf_series, num_classes).bins.tolist()
    if classifier == 'Equal_Interval':
        threshold_scale=ps.esda.mapclassify.Equal_Interval(gdf_series, num_classes).bins.tolist()
    if classifier == 'Quantiles':
        threshold_scale=ps.esda.mapclassify.Quantiles(gdf_series, num_classes).bins.tolist()
    if classifier == 'SplitSix':
        threshold_scale=split_six(gdf_series)

    return threshold_scale
    
threshold_scale = split_six(mtct3_series_df['TRpct'])
threshold_scale = get_threshold_scale('SplitSix', 6, mtct3_series_df['TRpct'])
```

```python
# TODO: Figure out how to combine the following into an interactive map

m2013 = folium.Map(width=500,height=500, location=[13, 101], zoom_start=5, tiles="cartodbpositron")
m2013.choropleth(
    geo_data=geo_data1,
    data=mtct3_series_df[mtct3_series_df['year']==2013],
    columns=['ISO_CODE', 'TRpct'],
    key_on='feature.properties.ISO_CODE',
    legend_name='Transmission Rate (%)',
    fill_color='RdPu',
    fill_opacity=0.7,
    line_color='black',
    line_weight=0.3,
    threshold_scale=threshold_scale,
    highlight=True)

m2014 = folium.Map(width=500,height=500, location=[13, 101], zoom_start=5, tiles="cartodbpositron")
m2014.choropleth(
    geo_data=geo_data1,
    data=mtct3_series_df[mtct3_series_df['year']==2014],
    columns=['ISO_CODE', 'TRpct'],
    key_on='feature.properties.ISO_CODE',
    legend_name='Transmission Rate (%)',
    fill_color='RdPu',
    fill_opacity=0.7,
    line_color='black',
    line_weight=0.3,
    threshold_scale=threshold_scale,
    highlight=True)

m2015 = folium.Map(width=500,height=500, location=[13, 101], zoom_start=5, tiles="cartodbpositron")
m2015.choropleth(
    geo_data=geo_data1,
    data=mtct3_series_df[mtct3_series_df['year']==2015],
    columns=['ISO_CODE', 'TRpct'],
    key_on='feature.properties.ISO_CODE',
    legend_name='Transmission Rate (%)',
    fill_color='RdPu',
    fill_opacity=0.7,
    line_color='black',
    line_weight=0.3,
    threshold_scale=threshold_scale,
    highlight=True)

m2016 = folium.Map(width=500,height=500, location=[13, 101], zoom_start=5, tiles="cartodbpositron")
m2016.choropleth(
    geo_data=geo_data1,
    data=mtct3_series_df[mtct3_series_df['year']==2016],
    columns=['ISO_CODE', 'TRpct'],
    key_on='feature.properties.ISO_CODE',
    legend_name='Transmission Rate (%)',
    fill_color='RdPu',
    fill_opacity=0.7,
    line_color='black',
    line_weight=0.3,
    threshold_scale=threshold_scale,
    highlight=True)
```
